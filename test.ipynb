```python
import random
import matplotlib.pyplot as plt

def roll_2d6(n_times):
    results = []
    for _ in range(n_times):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        results.append(die1 + die2)
    return results
    

# Roll the dice 100000 times
roll_results = roll_2d6(100000)

#Plot results
plt.hist(roll_results, bins = range(2,14), align='left', rwidth=0.9)
plt.title('Distribution of 100000 2d6 rolls')
plt.xlabel('Sum of two dice')
plt.ylabel('Frequency')
plt.xticks(range(2,13))
plt.show()


# another test comment!
```
